```python
#dealing with missing data

import missingno as msno
import pandas as pd

msno.matrix(data)
print("Percentage of missing values:")
print(data.isnull().mean() * 100)
```

```python
msno.bar(data)
```

```python
threshold = 0
n_samples = len(data)
missing_rate=data.isnull().sum()/n_samples * 100
print("list of features with missing rate higher than {}%.\n{}".format(threshold, missing_rate[missing_rate>threshold]))
```

```python
threshold = 0
n_missing = data.isnull().sum()
print("list of features with missing rate higher than {}%.\n{}".format(threshold, n_missing[n_missing>threshold]))
```

```python
df_missing = pd.concat([n_missing, missing_rate], axis=1)
df_missing = df_missing.rename(columns={0: 'Total', 1: 'missing_ratio'})
df_missing = df_missing.query('Total>0 and missing_ratio>0')
display(df_missing)
```

```python
threshold =5
five_missing_features = df_missing.query('Total >= {}'.format(threshold)).index
print("list of features with {} or more missing: \n{}".format(threshold, five_missing_features))
```

```python
df_dropped_features = data.drop(five_missing_features, axis=1)
display(df_dropped_features.head())
```

```python
df_dropna = df_dropped_features.dropna(axis=0)
df_dropna = df_dropna.reset_index(drop=True)
print('Number of samples remaining: {}'. format(len(df_dropna)))
msno.bar(df_dropna)
```

# PROBLEM 5

Kurtosis:

Kurtosis is a statistical measure that describes the distribution of data in a dataset. It measures the "tailedness" or the degree of outliers in a distribution.
A distribution with high kurtosis has heavy tails or outliers, indicating a high probability of extreme values. On the other hand, a distribution with low kurtosis has light tails, indicating fewer outliers.
There are two common types of kurtosis: positive (leptokurtic) and negative (platykurtic). Positive kurtosis means the distribution has fatter tails than a normal distribution, while negative kurtosis means the distribution has thinner tails.

Skewness:

Skewness is a measure of the asymmetry of a probability distribution. It indicates whether the data is concentrated on one side of the mean more than the other.
A positive skewness indicates that the right tail of the distribution is longer or fatter than the left tail, meaning there are more outliers on the right side. Conversely, a negative skewness means that the left tail is longer or fatter than the right tail.
Skewness helps identify the direction and relative magnitude of skew (asymmetry) in a dataset. Unlike kurtosis, skewness does not differentiate between the tails.

```python
import seaborn as sns
import numpy as np
import pandas as pd

target = data['SalePrice']
print("Skewness= {: .2f}  sharpness={: .2f}".format(target.skew(), target.kurtosis()))
display(sns.displot(target, kde=False, rug=False, bins=10))
```

```python
#Logrithm transmission
target_log = target.apply(np.log)

print("Skewness= {: .2f}  sharpness={: .2f}".format(target_log.skew(), target_log.kurtosis()))
display(sns.displot(target_log, kde=False, rug=False, bins=10))
```

```python
# Identify and exclude non-numeric columns
numeric_columns = data.select_dtypes(include=[np.number]).columns
numeric_data = data[numeric_columns]

# Create the correlation matrix
df_corr = numeric_data.corr()

colormap= plt.cm.RdBu
plt.subplots(figsize=(15, 12))
plt.title('Correlation Matrix')
sns.heatmap(df_corr, linewidths=0.1, vmax=1.0,
           square=True, cmap=colormap, linecolor='white', annot=False)
plt.show()
```

```python
#sort the correlation with Saleprice

df_corr_saleprice = df_corr['SalePrice'].sort_values(ascending=False)
print(df_corr_saleprice[1:11])
```

```python
df_high_corr_saleprice  = data[df_corr_saleprice[1:11].index]
display(df_high_corr_saleprice.head())

df_high_corr_saleprice_corr = df_high_corr_saleprice.corr()

colormap = plt.cm.RdBu
plt.subplots(figsize=(7,7))
plt.title('Correlation matrix')
sns.heatmap(df_high_corr_saleprice_corr, linewidths=0.1, vmax=1.0,
           square=True, cmap=colormap, linecolor='white', annot=False)
plt.show()
```
